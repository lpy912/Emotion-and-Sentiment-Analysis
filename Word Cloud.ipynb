```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import nltk
from nltk import pos_tag_sents
```

```python
import spacy
```

```python
#df = pd.read_csv("boostonly.csv", index_col=0)
df = pd.read_csv("tngonly.csv", index_col=0)
#df = pd.read_csv("mbbonly.csv", index_col=0)
```

```python
df.head()
```

```python
# Start with one review:
text = " ".join(review for review in df.Review)
```

```python
tokens=nltk.word_tokenize(text)
```

```python
out=nltk.pos_tag(tokens)
```

```python
POSreview=pd.DataFrame(out)
```

```python
#filter by POS needed here
POSfilter=POSreview[(POSreview[1]=='JJ')]
```

```python
#join all words again
text2 = " ".join(review for review in POSfilter[0])
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["app", "boost","App","TnG","tng","apps","use","touch","go","will","wallet","ewallet","please","card"])
```

```python
# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text2)
```

```python
# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
